# **Statistics Basic Assignment**

1. What is statistics, and why is it important?
  - Statistics is the study of collecting, analyzing, interpreting, and presenting data. It helps in making informed decisions using data patterns and trends.
2.  What are the two main types of statistics?
  - The two main types are descriptive statistics, which summarize data, and inferential statistics, which make predictions based on data samples.
3. What are descriptive statistics?
  - Descriptive statistics describe and summarize features of a dataset using measures like mean, median, mode, range, and standard deviation.
4. What is inferential statistics?
  - Inferential statistics involve making predictions or inferences about a population based on data sampled from it.
5. What is sampling in statistics?
  - Sampling is the process of selecting a subset of individuals from a population to estimate characteristics of the whole population.
6. What are the different types of sampling methods?
  - Sampling methods include random, systematic, stratified, cluster, convenience, and judgmental sampling.
7. What is the difference between random and non-random sampling?
  - Random sampling gives each member a fair chance to be selected, while non-random sampling relies on judgment or convenience, which may cause bias.
8. Define and give examples of qualitative and quantitative data.
  - Qualitative data is non-numeric (e.g., colors, gender), while quantitative data is numeric (e.g., height, age).
9. What are the different types of data in statistics?
  - Data types include nominal, ordinal, interval, and ratio, based on the nature and measurement of values.
10. Explain nominal, ordinal, interval, and ratio levels of measurement.
  - Nominal: Categories (e.g., gender).Ordinal: Ranked order (e.g., class levels).Interval: Numeric with no true zero (e.g., temperature).Ratio: Numeric with true zero (e.g., weight).
11. What is the measure of central tendency?
  - It refers to values that represent the center or average of a datasetâ€”mainly mean, median, and mode.
12. Define mean, median, and mode.
  - Mean: Average value.
  - Median: Middle value when sorted.
  - Mode: Most frequent value.
13. What is the significance of the measure of central tendency?
  - It helps summarize large datasets and provides insight into the most typical or central values in the data.
14. What is variance, and how is it calculated?
  - Variance measures the spread of data from the mean. It is calculated by averaging the squared differences from the mean.
15. What is standard deviation, and why is it important?
  - Standard deviation shows how much data varies from the mean. It is important for understanding data consistency.
16. Define and explain the term range in statistics.
  - The range is the difference between the maximum and minimum values in a dataset. It shows the spread of data.
17. What is the difference between variance and standard deviation?
  - Both measure data spread, but standard deviation is the square root of variance and is in the same unit as the data.
18. What is skewness in a dataset?
  - Skewness measures the asymmetry of the data distribution. It shows whether the data leans more to the left or right.
19. What does it mean if a dataset is positively or negatively skewed?
  - Positively skewed: Tail is on the right; more low values.
  - Negatively skewed: Tail is on the left; more high values.
20. Define and explain kurtosis.
  - Kurtosis measures the "tailedness" of a distribution. High kurtosis indicates heavy tails; low kurtosis indicates light tails.
21. What is the purpose of covariance?
  - Covariance shows how two variables change together. A positive value means they increase together; a negative means opposite.
22. What does correlation measure in statistics?
  - Correlation measures the strength and direction of the relationship between two variables, typically on a scale from -1 to 1.
23. What is the difference between covariance and correlation?
  - Covariance shows direction but not scale, while correlation standardizes it, showing both direction and strength of a relationship.
24. What are some real-world applications of statistics?
  - Statistics is used in healthcare, business forecasting, quality control, economics, machine learning, and public policy decisions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler

# 1. Mean, Median, Mode
data = [12, 15, 12, 18, 19, 21, 18, 18, 20]
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Mode:", stats.mode(data, keepdims=False).mode)

# 2. Variance and Standard Deviation
print("Variance:", np.var(data))
print("Standard Deviation:", np.std(data))

# 3. Data Classification
nominal = ['Red', 'Blue', 'Green']
ordinal = ['Low', 'Medium', 'High']
interval = [10, 20, 30]
ratio = [5.0, 10.0, 15.0]
print("Nominal:", nominal)
print("Ordinal:", ordinal)
print("Interval:", interval)
print("Ratio:", ratio)

# 4. Sampling Techniques
population = list(range(1, 101))
random_sample = np.random.choice(population, 10, replace=False)
df = pd.DataFrame({'Group': ['A']*50 + ['B']*50, 'Value': np.random.randint(1, 100, 100)})
stratified_sample = df.groupby('Group', group_keys=False).apply(lambda x: x.sample(5))
print("Random Sample:", random_sample)
print("Stratified Sample:\n", stratified_sample)

# 5. Calculate Range
def calculate_range(data):
    return max(data) - min(data)
print("Range:", calculate_range(data))

# 6. Histogram to visualize skewness
plt.hist(data, bins=5)
plt.title("Histogram")
plt.show()

# 7. Skewness and Kurtosis
print("Skewness:", stats.skew(data))
print("Kurtosis:", stats.kurtosis(data))

# 8. Demonstrate skewness
positive_skew = np.random.exponential(scale=2, size=1000)
negative_skew = -np.random.exponential(scale=2, size=1000)
sns.histplot(positive_skew, kde=True)
plt.title("Positive Skew")
plt.show()
sns.histplot(negative_skew, kde=True)
plt.title("Negative Skew")
plt.show()

# 9. Covariance
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
covariance = np.cov(x, y)[0][1]
print("Covariance:", covariance)

# 10. Correlation
correlation = np.corrcoef(x, y)[0][1]
print("Correlation Coefficient:", correlation)

# 11. Scatter plot
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# 12. Random vs Systematic Sampling
systematic_sample = population[::10]
print("Systematic Sample:", systematic_sample)

# 13. Grouped Data: Mean, Median, Mode
grouped_data = pd.DataFrame({'Class Interval': ['10-20', '20-30', '30-40'], 'Frequency': [5, 8, 7]})
grouped_mean = (15*5 + 25*8 + 35*7) / sum([5,8,7])
print("Grouped Mean:", grouped_mean)

# 14. Simulated Data
sim_data = np.random.normal(loc=50, scale=10, size=100)
print("Simulated Mean:", np.mean(sim_data))
print("Simulated Std Dev:", np.std(sim_data))

# 15. Descriptive Stats
print("Descriptive Statistics:\n", pd.Series(sim_data).describe())

# 16. Boxplot
sns.boxplot(x=sim_data)
plt.title("Boxplot")
plt.show()

# 17. IQR
Q1 = np.percentile(sim_data, 25)
Q3 = np.percentile(sim_data, 75)
print("IQR:", Q3 - Q1)

# 18. Z-Score Normalization
scaler = StandardScaler()
z_scores = scaler.fit_transform(sim_data.reshape(-1, 1))
print("Z-scores (first 5):", z_scores[:5].flatten())

# 19. Compare Two Datasets Std Dev
data1 = np.random.normal(30, 5, 100)
data2 = np.random.normal(30, 15, 100)
print("Std Dev Data1:", np.std(data1))
print("Std Dev Data2:", np.std(data2))

# 20. Covariance Heatmap
df_cov = pd.DataFrame({'x': x, 'y': y})
sns.heatmap(df_cov.cov(), annot=True, cmap='coolwarm')
plt.title("Covariance Heatmap")
plt.show()

# 21. Correlation Matrix
df_corr = pd.DataFrame(np.random.randn(100, 3), columns=['A', 'B', 'C'])
sns.heatmap(df_corr.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# 22. Variance & Std Dev Again
data3 = np.random.randint(10, 50, 50)
print("Variance:", np.var(data3))
print("Std Dev:", np.std(data3))

# 23. Visualize Skewness & Kurtosis
sns.histplot(sim_data, kde=True)
plt.title(f"Skew: {stats.skew(sim_data):.2f}, Kurtosis: {stats.kurtosis(sim_data):.2f}")
plt.show()

# 24. Pearson & Spearman Correlation
pearson_corr = stats.pearsonr(x, y)[0]
spearman_corr = stats.spearmanr(x, y)[0]
print("Pearson Correlation:", pearson_corr)
print("Spearman Correlation:", spearman_corr)
```
